```python
college.describe()  # Display summary statistics of the dataframe
```

```python
pd.plotting.scatter_matrix(college[['Enroll', 'Top10perc', 'Accept']], figsize=(10, 10), diagonal='kde')
```

**Explaination**

Enroll vs Enroll: Most of the school looks like to enroll 1000 where the distribution is peak => Right Skewed

Top10perc vs Top10perc: Most of the school have 20-40% of top students where the distribution is peak

Accept vs Accept: Most of the school accept around 2500 students => Right Skewed

Enroll vs Accept: Strong positive relations, more accepted students = more enrolled student

Top10perc vs Accept/Enroll: No clear relationship, more Top10perc != more accept/enroll

```python
college.boxplot(column='Outstate', by='Private', figsize=(10, 5), vert=True)
```

```python
college['Elite'] = pd.cut(college['Top10perc'], [0,50,100], labels=['No', 'Yes'])
college.boxplot(column='Outstate', by='Elite', figsize=(10, 5), vert=True)
```

```python
college['Elite'].value_counts()  # Count the number of colleges in each category of 'Elite'
```

```python
# plot histograms for various columns into 2*2 grid
import matplotlib.pyplot as plt
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

college['Outstate'].plot.hist(bins=20, edgecolor='black', alpha=0.7, ax=axs[0, 0])
college['Enroll'].plot.hist(bins=20, edgecolor='black', alpha=0.7, ax=axs[0, 1])
college['Top10perc'].plot.hist(bins=20, edgecolor='black', alpha=0.7, ax=axs[1, 0])
college['Accept'].plot.hist(bins=20, edgecolor='black', alpha=0.7, ax=axs[1, 1])
```

```python
import pandas as pd
auto = pd.read_csv('./datasets/Auto.csv', na_values='?')
auto = auto.dropna()  # Drop rows with missing values
```

```python
auto.head()  # Display the first few rows of the Auto dataset
```

```python
# Find the quantitive and qualitative variables in the Auto dataset
quantitative_vars = auto.select_dtypes(include=['float64', 'int64']).columns.tolist
qualitative_vars = auto.select_dtypes(include=['object']).columns.tolist
quantitative_vars, qualitative_vars
```

```python
# range of values for quantitative variables
quantitative_ranges = {var: (auto[var].min(), auto[var].max()) for var in quantitative_vars()}
quantitative_ranges
```

```python
quantitative_mean = {var: auto[var].mean() for var in quantitative_vars()}
quantitative_std = {var: auto[var].std() for var in quantitative_vars()}
quantitative_mean, quantitative_std
```

```python
# Remove 10th to 85th observations from the Auto dataset
auto_trimmed = auto.drop(auto.index[9:85])
```

```python
auto_trimmed.describe()  # Display summary statistics of the trimmed Auto dataset
```
